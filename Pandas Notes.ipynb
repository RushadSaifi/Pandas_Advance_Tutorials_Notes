
# Data Manipulation

There are five types of Data manipulation task in the pandas which are:-
    1)Selecting/Indexing.
    2)Filtering.
    3)Sorting.
    4)Mutuating/Conditionally adding columns
    5)Group by/Summarize.

# Selecting/ Indexing 

There are two attributes for Selecting and Indexing. which are:- 
    iloc & loc

Note:- In the Selection task there is no effect on original DataFrame.

iloc:-    This is used to print the selcted part of the dataframe as we seen in the below example. here first argument is used for the row and second argument is used for the columns. and in below example we put range like 0:3 means from 0 to 2(last  excluded in range) & 1:3 means 1 to 2.

```python
df.iloc[0:3, 1:3]
```

Select Specific columns without range using iloc. here scond argument is a list and we put the index of the specific column which we want to access.

```python
df.iloc[0:5, [0,3]]
```

Select all the columns and ranges in rows.

```python
df.iloc[0:5]
```

Select all the columns and all the rows.

```python
df.iloc[:,:]
```

Select all rows and all columns except last column

```python
df.iloc[:,:-1]
```

May be iloc will be depricated soon so there is a new concept advance of iloc which is loc.

Selection with the help of loc:-

All rows and specific columns

```python
df.loc[:, ["Country"]]
```

All rows and multiple Columns

```python
df.loc[:, ['Country', 'Purchased']]
```

All rows &  all Columns

```python
df.loc[:,:]
```

# Filtering

This is used to filter the DataFrame.

Simple Filtering:- in this type of filtering only boolean result is out as you seen in the result.

```python
df['Age']>30
```

Condtional filtering. :- if the condition is true then the value is added in the column as shown in example below.

```python
df.loc[df['Age']>30, ['Age']]
```

Add more than one Column in conditional filtering.

```python
df.loc[df['Age']>30, ['Age','Country', 'Purchased']]
```

You can Filtering the dataframe without loc too. as shown below.

```python
df[df['Age']>30]
```

.isin() method:- this is used to filter the dataframe where specific value is present. as hown in the below example it filters those rows where only spain and france are exist.

```python
df[df['Country'].isin(['Spain', 'France'])]
```

We can add the conditonal colum as shown in the example below but with the help of numpy library.

#  Grouping

We can group the same value in a specific column and find their mean, mode, median, sum, etc.

```python
df.groupby('Country')['Age'].agg("mean")
```

```python
df.groupby('Country')['Age'].agg('sum')
```

# Wroking with NAn Values 

It will drop the col and row in which signle nan is present

```python
y.dropna(axis=0)
```

```python
y.dropna(axis=1)
```

Drop row if all the value in row is nan

```python
y.dropna(how='all')
```

Drop Col & row if the col and row has greater no. of nan values than threshold.

```python
y.dropna(axis=1, thresh=4)
```

Fill Na Values

```python
y.fillna(0)
```

```python
y.fillna(1)
```

```python
y.fillna({'A':4, 'B':2, 'C':1}) #fill na values for specific columna with the help of dictionary.
```

```python
y.fillna(y.mean()) # fill na values with the mean of the column. column must be integer datatypes.
```

```python
y.fillna(method='ffill') # fill nan value forward 
```

```python
y.fillna(method='bfill') # fill nan values backward.
```
